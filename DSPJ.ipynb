```python
# 필요모듈 import
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#df 불러오기
df = pd.read_csv('./dcpj.csv')
df
```

```python
#교육 카테고리 조건 
edu_cond = (df.category_title == '교육')
```

```python
#교육 카테고리만 추출
df_edu = df.loc[edu_cond]
```

```python
#교옥 카테고리 확인
df_edu.info()
```

```python
#20개 정도 확인한다.
df_edu.head(20)
```

```python
#customer id로 그룹화, 각 고객별 type을 알아보기 위해서
df_edu_groupby_customer_id = df_edu.groupby('customer_id').count()
```

```python
#type이 1개가 아닌 customer_id를 살펴보는 조건
type_edu_cond = df_edu_groupby_customer_id['type'] != 1
```

```python
#조건 적용해서 customer_id를 확인할 수 있다.
df_edu_groupby_customer_id_multiple = df_edu_groupby_customer_id.loc[type_edu_cond]
```

```python
# type 개수가 1보다 큰 자료들 customer_id로 추출
df_edu_groupby_customer_id_multiple
```

```python
#index인 customer_id를 뽑아서 리스트로 만든다.
df_edu_groupby_customer_id_multiple.index.values.tolist()
```

```python
#type이 여러개인 customer_id들이 포함된 리스트에 속하는지 아닌지 Boolean 형으로 반환하는 조건
multiple_type_cond = df_edu['customer_id'].isin(df_edu_groupby_customer_id_multiple.index.values.tolist())
```

```python
#type이 여러개이며 교육category에 속하는 자료 뽑음
df_edu.loc[multiple_type_cond]
```
